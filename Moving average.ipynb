```python
import pandas as pd
import numpy as np
import pickle
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
t = np.arange(0,2,0.001)
X = np.sin(2*np.pi*t) + np.sin(3* 2*np.pi*t) + np.sin(5* 2*np.pi*t) + np.sin(7* 2*np.pi*t) + np.sin(9* 2*np.pi*t)
plt.plot(t, X)
```

```python
# Moving average = pandas
df = pd.DataFrame(X)
ma = df.rolling(window=100)
ma_mean = ma.mean()
df.plot()
ma_mean.plot(color='red')
```
